Cross sectional data - информация об N наблюдениях в один период времени $(1*N)$ \
Time series data - информация об одном наблюднении за T периодов времени $(T*1)$ \
Panel data - информация об N наблюдениях за T периодов $(T*N)$

```python
#Time Series data structure
ts_life_exp_RUS = life_exp[life_exp['CountryCode'] == 'RUS'][['ValueYear', 'LifeExpectancy']]
ts_life_exp_RUS.reset_index(drop = True, inplace = True)
ts_life_exp_RUS
```

Сделайте самостоятельно Cross sectional data structure за 2017 год по всем N странам

```python
cs_life_exp_2017 = life_exp[life_exp['ValueYear'] == 2017][['CountryCode', 'LifeExpectancy']]
cs_life_exp_2017.reset_index(drop = True, inplace = True)
cs_life_exp_2017
```

Pivot table - сводная таблица в нашем случае, она же panel data формат

```python
pd.pivot_table(life_exp, values = 'LifeExpectancy', columns = 'ValueYear', index = 'CountryCode')
```

```python
panel = pd.pivot_table(life_exp, values = 'LifeExpectancy', index = 'ValueYear', columns = 'CountryCode')
```

```python
Merge - Concat - Join
```

```python
life_exp_ext = pd.merge(life_exp, cntr[['CountryCode','Name','Continent']], how = 'inner', on = 'CountryCode')
```

```python
life_exp_ext[life_exp_ext['ValueYear'] == 2017].groupby(by = ['Continent'])['LifeExpectancy'].mean()
```

```python
pan1 = panel[:len(panel)//2]
pan2 = panel[len(panel)//2:]
```

```python
pd.concat([pan1, pan2])
```
